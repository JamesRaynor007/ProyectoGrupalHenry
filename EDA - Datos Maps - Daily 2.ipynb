```python
def limpiar_caracteres_extraños(texto):
    if isinstance(texto, str):  # Verificamos que el texto sea una cadena
        return ''.join(e for e in texto if e.isalnum() or e.isspace())
    return texto  # Retornamos el texto original si no es una cadena

# Aplicar la limpieza a todas las columnas de tipo string
for col in Prueba.select_dtypes(include=['object']):
    Prueba[col] = Prueba[col].apply(limpiar_caracteres_extraños)
```
